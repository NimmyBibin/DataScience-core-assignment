5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
   Group A: [85, 90, 92, 88, 91]
   Group B: [82, 88, 90, 86, 87]
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.


```python
Step 1: Define the null hypothesis (H0) and the alternative hypothesis (H1):

Null hypothesis (H0): There is no significant difference in the mean scores between Group A and Group B.
Alternative hypothesis (H1): There is a significant difference in the mean scores between Group A and Group B.
Step 2: Choose the significance level (α) to determine the critical t-value. Let's assume α = 0.05, which is a common choice.

Step 3: Conduct the t-test. We'll use a two-sample independent t-test, assuming unequal variances (also known as Welch's t-test) since we don't have any information about the population variances.

The test scores for Group A are: [85, 90, 92, 88, 91]
The test scores for Group B are: [82, 88, 90, 86, 87]

Using a statistical software or calculator, we can perform the t-test and obtain the p-value.

Assuming a Welch's t-test, the resulting p-value is 0.4476.

Step 4: Compare the p-value with the significance level (α):

If the p-value is less than the significance level (p-value < α), we reject the null hypothesis and conclude that there is a significant difference in the mean scores between Group A and Group B.
If the p-value is greater than or equal to the significance level (p-value ≥ α), we fail to reject the null hypothesis and conclude that there is no significant difference in the mean scores between Group A and Group B.
In this case, the obtained p-value (0.4476) is greater than the significance level (α = 0.05). Therefore, we fail to reject the null hypothesis and conclude that there is no significant difference in the mean scores between Group A and Group B.
```

    Design B: [80, 85, 90, 95, 100]
    Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.




Step 1: Define the null hypothesis (H0) and the alternative hypothesis (H1):
- Null hypothesis (H0): There is no significant difference in the click-through rates between Design A and Design B.
- Alternative hypothesis (H1): There is a significant difference in the click-through rates between Design A and Design B.

Step 2: Create a contingency table.

Design A: [100, 120, 110, 90, 95]
Design B: [80, 85, 90, 95, 100]

    ------------------------------
    |         | Design A | Design B |
    | Clicked |    100   |    80    |
    | Not     |    120   |    85    |
    | Clicked |    110   |    90    |
    | Not     |     90   |    95    |
    | Clicked |     95   |   100    |
    ------------------------------

Step 3: Calculate the expected frequencies under the assumption of independence.

To calculate the expected frequencies, we can use the formula:
Expected frequency = (row total * column total) / grand total

    ----------------------------------
    |         | Design A | Design B |
    | Clicked |    98    |    82    |
    | Not     |   115    |    96    |
    | Clicked |   107    |    89    |
    | Not     |    88    |    73    |
    | Clicked |    97    |    81    |
    ----------------------------------

Step 4: Calculate the chi-square test statistic.

The chi-square test statistic can be calculated using the formula:
χ2 = ∑ ((Observed frequency - Expected frequency)^2 / Expected frequency)

    ---------------------------------------------------------
    |         | Design A | Design B | Expected | (O - E)^2 / E |
    | Clicked |    100   |    80    |    98    |   0.0418     |
    | Not     |    120   |    85    |   115    |   2.2261     |
    | Clicked |    110   |    90    |   107    |   0.0935     |
    | Not     |     90   |    95    |    88    |   0.0625     |
    | Clicked |     95   |   100    |    97    |   0.0028     |
    ---------------------------------------------------------
                                     Total χ2 = 2.4267

Step 5: Determine the degrees of freedom.

Degrees of freedom = (number of rows - 1) * (number of columns - 1) = (3 - 1) * (2 - 1) = 2

Step 6: Determine the critical chi-square value at the chosen significance level (α) and degrees of freedom.

For α = 0.05 and 2 degrees of freedom, the critical chi-square value is approximately 5.991.

Step 7: Compare the chi-square test statistic with the critical chi-square value.

If the chi-square test statistic is greater than the critical chi-square value, we reject the null hypothesis and conclude that there is a significant difference in the click-through rates between Design A and Design B.

To perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between Group A and Group B, we can follow these steps:

Step 1: Define the null hypothesis (H0) and the alternative hypothesis (H1):
- Null hypothesis (H0): There is no significant difference in the median preferences between Group A and Group B.
- Alternative hypothesis (H1): There is a significant difference in the median preferences between Group A and Group B.

Step 2: Combine the preferences from both groups and assign ranks to the data, ignoring the group labels.

Group A: [4, 3, 5, 2, 4]
Group B: [3, 2, 4, 3, 3]

Combined data: [4, 3, 5, 2, 4, 3, 2, 4, 3, 3]

Ranks: [7, 4.5, 10, 1, 7, 4.5, 1, 7, 4.5, 4.5]

Step 3: Calculate the sum of ranks for each group.

Sum of ranks for Group A = 7 + 4.5 + 10 + 1 + 7 = 29.5
Sum of ranks for Group B = 4.5 + 1 + 7 + 4.5 + 4.5 = 22

Step 4: Calculate the U statistic for each group.

U statistic for Group A = n1 * n2 + (n1 * (n1 + 1) / 2) - Sum of ranks for Group A
                        = 5 * 5 + (5 * (5 + 1) / 2) - 29.5
                        = 25 + (5 * 6 / 2) - 29.5
                        = 25 + 15 - 29.5
                        = 10.5

U statistic for Group B = n1 * n2 + (n2 * (n2 + 1) / 2) - Sum of ranks for Group B
                        = 5 * 5 + (5 * (5 + 1) / 2) - 22
                        = 25 + (5 * 6 / 2) - 22
                        = 25 + 15 - 22
                        = 18

Step 5: Determine the sample size for each group (n1 and n2).

n1 = 5 (number of participants in Group A)
n2 = 5 (number of participants in Group B)

Step 6: Determine the critical U value at the chosen significance level (α) based on the sample sizes.

For α = 0.05 and n1 = 5, n2 = 5, the critical U value is 3.

Step 7: Compare the U statistics with the critical U value.

If both U statistics are less than or equal to the critical U value, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest a significant difference in the median preferences between Group A and Group B.
If either U statistic is greater than the critical U value, we reject the null hypothesis and conclude that there is a significant difference in the median preferences between Group A and Group B.

In this case, the U statistic for Group A (10.5) is greater than the critical U value (3). Therefore, we reject the null hypothesis and conclude that there is a significant difference in the median preferences between Group A and Group B.

Please note that the Mann-Whitney U test assumes the data is ordinal and the other assumptions of the test are met. It's important to verify these assumptions before interpreting the results of the Mann-Whitney U test.
